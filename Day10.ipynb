```python
## Python List: In Python, lists allow us to store multiple items in a single variable. For example, if you need to store the ages of all the students in a class, you can do this task using a list.
## lists are similar to arrays(dynamic arrays that allow us to store items of different data types) in other programming languages.

names=["aksh","anil","manu","mallu","sammu","chinnu"]
print(names)
print(type(names))
```

```python
## List Characteristics: IN Python,lists are:
#1.Ordered-They maintain the order of elements.
#2.Mutable-Items can be changed after creation.
#3.Allow duplicates-They can contain duplicate values.

names=["aksh","anil","manu","mallu","sammu","chinnu"]     # indexing
print(names[0])
names[0]="Akshata"
print(names[0])
print(names[4])
print(names[-1])
print(len(names))
```

```python
## Operations in list: len(),Negative Indexing,append(),extend(),insert(),remove(),pop(),clear(),min() max(),slicing,count(),sort(),reverse(),copy(),index()

list1=[4,32,4,6,2,7]
list1.append(9)        # adding value at the end is append()
list1.append([1,2,3])
list1.extend([1,2,3])    # in extend() elements are separately adding at the end
print(list1)
```

```python
# insert(): is inserting the value at index point

list1=[1,2,3,4,5,6]
list1.insert(3,9)

print(list1)
```

```python
# remove(): specific element remove and pop():specific index remove or last element in the list remove

list1=[1,2,3,4,5,6]
list1.remove(4)
list1.pop()

print(list1)
```

```python
list1=[1,2,3,4,5,6]
list1.remove(4)
list1.pop()
print(list1)

list1.clear()
print(list1)
```

```python
list1=[1,2,3,4,5,6]
print(min(list1))
print(max(list1))
```

```python
list1=[4,1,2,4,53,2,4,5,4,6,8,3,4,46]
print(list1.count(4))
```

```python
list1=[4,1,2,4,53,2,4,5,4,6,8,3,4,46]
list1.sort()    # O(n log n)
print(list1)
```

```python
list1=[4,1,2,4,53,2,4,5,4,6,8,3,4,46]
list1.reverse()    # O(n)
print(list1)
```

```python
list1=[1,5,3,2,5,7,3]

print(list1.index(2))
print(list1.index(5))
print(list1.index(5,2))
```

```python
list1=[1,2,3]
list2=list1   # deepcopy

list2.append(4)

print(list1)
print(list2)

print(list1,id(list1))
print(list2,id(list2))
```

```python
list1=[1,2,3]
list2=list1.copy()   # shallow copy

list2.append(4)

print(list1)
print(list2)

print(list1,id(list1))
print(list2,id(list2))
```

```python
## Slicing:

list1=[2,33,43,156,7,42,5,73,63]
print(list1[2:5])
print(list1[2:8:3])
print(list1[8:1:-2])

print(list1[:5])
print(list1[6:])
print(list1[:])
```

```python
list1=[3,4,456,24,312,234,456,42,23]

print(list1[::-1])
```
