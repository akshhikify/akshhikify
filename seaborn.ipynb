```python
## Correlation with heatmap::: A coorelation heatmap uses colored cells, typically in a monochromatic scale,to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in future Selection .It is used in exploratory data analysis
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True)) #heatmap actually helps us to find out the visualization stuf
```

```python
## joinlpot: A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation.It is usually a scatterplot, a hexbin plot,a 2D histogram or a 2D density plot.
```

```python
## Univariate analysis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
plt.show()
```

```python
## Pair plot : A "pair plot" is also known as scatterplot, in which one of the same data row is matched with another variable's value,like this .Pairs plots are just elaborations on this, showing all variables paired with all the other variables.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("tips")
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df,hue='sex')
```

```python
## Dist plot
# dist plot helps us to check the distribution of the columns feature
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("tips")
sns.distplot(df['tip'],kde=True)
plt.show()
```

```python
sns.distplot(df['tip'],kde=False,bins=10)     # kde= karnal density estimation
```
