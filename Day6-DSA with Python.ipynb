```python
## Time Complexity: Time complexity in computer science describes how the execution time of an algorithm grows as the input size increases.It essentially quantifies the number of operation an algorithm performs as a function of the input data size.
```

```python
## Asymptotic Notations: 1.Best Case 2.Average Case 3.Worst Case

## Types of Time complexities : Time complexities are categorized based on how an algorithm's runtime scales with the input size.Common types include constant time (o(1)), logarithmic time (O(log n)), linear time(O(n)), quadratic time(O(n^2)), and exponential time (O(2^n))
```

```python
## Linear Complexity
n=10
for i in range(n//2):
    print(i, end=" ")
```

```python
## Constant time (O(1))

for i in range(10):
    print(i, end=" ")
```
