## Functions in Python

```python
def even_odd(num):
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
```

```python
even_odd(24)
```

```python
# print vs return
def hello_world():
    print("Hello Welcome")
```

```python
val=hello_world()
```

```python
print(val)
```

```python
# 1st is positional argument
# 2nd is keyword argument
def hello(name,age=29):
    print("My name is {} and age is {}".format(name,age))
```

```python
hello("aksh")
```

```python
def hello(*args,**kwargs):
    print(args)
    print(kwargs)
```

```python
hello('krish','Naik',age=29,dob=1990)
```

```python
lst=['krish', 'Naik']
dict_args={'age':21, 'dob':2005}
```

```python
hello(*lst,**dict_args)
```

```python
lst=[1,2,3,4,5,6,7]
```

```python
def evenoddsum(lst):
    even_sum=0
    odd_sum=0
    for i in lst:
        if i%2==0:
            even_sum=even_sum+i
        else:
            odd_sum=odd_sum+i
    return even_sum,odd_sum
```

```python
evenoddsum(lst)
```
