```python
num=10
assert num>=10
```

```python
try:
    num=int(input("Enter a number: "))
    assert num%2==0
    print("The number is even ")
except AssertionError:
    print("Please enter even number")
```

```python
## Python generators vs iterators
```

```python
lst=[1,2,3,4]
for i in lst:
    print(i)
```

```python
iter(lst)
```

```python
## iterator
iterable=iter(lst)
```

```python
for i in iterable:
    print(i)
```

```python
iterable
```

```python
next(iterable)
```

```python
try:
    print(next(iterable))
except StopIteration:
    print("The iterator is empty")
```

```python
#### generator
def square(n):
    for i in range(n):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
### Difference
## 1. To create iterator we use iter() and to generator we use function along with yield keywords
## 2.generator uses the yield keyword.it save the local variables
## 3.generate in python helps us top write fast and compact code
## 4.python iterator is much more memory efficient
```
