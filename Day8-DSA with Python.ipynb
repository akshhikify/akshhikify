```python
## Factorial Numbers :

def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)

print(fact(4))
```

```python
## Recursive Tree : o(2n) is time complexity

class Solution:
    def fib(self, n: int) -> int:
        # base case
        if n==0 or n==1:
            return n
        # recursive case
        return self.fib(n-1) + self.fib(n-2)
```

```python
## Power of two :

class Solution:
    def isPowerOfTwo(self,n: int) -> bool:
        while n%2==0:
            n//=2
        return n==1
```

```python
# Power of two

def isPowerOfTwo(self,n: int)->bool:
    # base case
    if n<=0:
        return False
    if n==1:
        return True
    if n%2!=0:
        return False
    # recursive case
    return self.isPowerOfTwo(n//2)
```
