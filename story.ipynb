```python
import random

# Lists of story elements
characters = ["a brave knight", "a young girl", "an alien", "a robot", "a wizard"]
places = ["in a magical forest", "on a distant planet", "in a haunted house", "in a big city", "in a secret lab"]
problems = ["found a hidden treasure", "fought a dragon", "got lost", "discovered a secret", "saved the world"]
twists = ["but it was all a dream", "and became a hero", "but faced a big danger", "and changed forever", "and unlocked a mystery"]

def generate_story():
    character = random.choice(characters)
    place = random.choice(places)
    problem = random.choice(problems)
    twist = random.choice(twists)

    story = f"Once upon a time, {character} was {place}. One day, they {problem}, {twist}."
    return story

# Generate story
print(generate_story())
```
