## Lambda functions

```python
## anonymous functions
## A function with no name
```

```python
def addition(a,b):
    return a+b
```

```python
addition(4,5)
```

```python
addition=lambda a,b:a+b
```

```python
addition(12,14)
```

```python
def even(num):
    if num%2==0:
        return True
```

```python
even(24)
```

```python
even(13)
```

```python
def addition(x,y,z):
    return x+y+z
```

```python
addition(2,3,4)
```

```python
addition=lambda x,y,z:x+y+z
```

```python
addition(2,3,4)
```

## Map Function

```python
def even_or_odd(num):
    if num%2==0:
        return "even".format(num)
    else:
        return "odd".format(num)
```

```python
even_or_odd(24)
```

```python
lst=[1,2,3,4,5,6,7,8,9,24,56,78]
```

```python
map(even_or_odd,lst)
```

```python
list(map(even_or_odd,lst))
```

## Filter Function in Python

```python
def even(num):
    if num%2==0:
        return True
```

```python
lst=[1,2,3,4,5,6,7,8,9,0]
```

```python
list(filter(even,lst))
```

```python
filter(lambda num:num%2==0,lst)
```

```python
list(filter(lambda num:num%2==0,lst))
```

```python
list(map(lambda num:num%2==0,lst))
```
