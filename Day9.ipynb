```python
# GCD Of Two Numbers-euclidean algorithm : The Euclidean algorithm is a method for efficiently calculating the greatest common divisor(GCD) of two integers. OR (HIGHEST COMMON FACTOR)
## LCM a*b=gcd(a,b)
# a=15=1,3,5,15
# b=50=1,2,5,20,25,50
# gcd(15,50)=5

def gcd(a,b):
    if b==0:
        return a
    return gcd(b,a%b)
def lcm(a,b):
    return a*b/gcd(a,b)
print(gcd(15,50))
print(lcm(15,50))
```

```python
class Solution:
    def findPow(x,n):
        # base case
        if n==0:
            return 1
        # recursive
        a = findpow(x,n//2)
        if n%2==0:
            return a*a
        else:
            return a*a*x
    def myPow(self, x: float, n: int)-> float:
        if n>=0:
            return self.findPow(x,n)
        else:
            return 1/self,findPow(x,n*(-1))
```
