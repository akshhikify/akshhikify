## List Comprehension

```python
# List comprehensions provide a concise way to create lists. It consists of brackets containing an expression followed by a for clause, then zero or more for or if clauses. The expression can put in all kind of objects in lists.
```

```python
lst1=[]
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
```

```python
lst_square([1,2,3,4,5,6,7])
```

```python
lst=[[1,2,3,4,5,6,7]]
```

```python
lst=[1,2,3,4,5,6,7]
squared=[i*i for i in lst]
print(squared)
```

```python
lst1=[i*i for i in lst if i%2==0]
```

```python
print(lst1)
```
