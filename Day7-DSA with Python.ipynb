```python
## Logarithmic Time (o(log n))

n=1000
while n>0:
    print(n, end=" ")
    n//=2
```

```python
i=1
n=100
while i<=n:
    print(i, end=" ")
    i*=2
```

```python
## Exponential  Time (O(2^n))
# Comparison of time Complexities
# O(1)<O(log n)<O(n)<O(n log n)<O(n^2)<O(2^n)<O(n!)
```
