```python
marks=int(input())
# >=90 Excellent
# 70-90 Good
# 40-70 Fair
# <40 Bad

if marks>=90:
    print("Excellent")
elif marks>=70:
    print("Good")
elif marks>=40:
    print("Fair")
else:
    print("Bad")
```

```python
temp = int(input())

# 25-50-> Hot
# 10-24-> Cold
# <10 -> Extremely cold

if temp<=50 and temp>=25:
    print("Hot")
elif temp<=24 and temp>=10:
    print("Cold")
elif temp<10:
    print("Extremely Cold")
```
